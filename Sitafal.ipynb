```python
!pip install pdfplumber
!pip install sentence-transformers
!pip install faiss-cpu
!pip install openai

from google.colab import files
import pdfplumber
import re
import numpy as np
import faiss
from sentence_transformers import SentenceTransformer

# Step 1: Upload PDF
uploaded = files.upload()

# Access the uploaded file
file_name = list(uploaded.keys())[0]
print("Uploaded file name:", file_name)

# Step 2: Extract text from the PDF
def extract_pdf_text(pdf_path, page_num=0):
    with pdfplumber.open(pdf_path) as pdf:
        page = pdf.pages[page_num]
        return page.extract_text()

# Extracting text from the first page
extracted_text = extract_pdf_text(file_name, page_num=0)
print("Extracted Text:", extracted_text[:500])  # Show first 500 characters

# Step 3: Chunking the text
def chunk_text(text, chunk_size=500):
    # Split the text into chunks of a specified size
    chunks = [text[i:i+chunk_size] for i in range(0, len(text), chunk_size)]
    return chunks

# Chunk the extracted text
chunks = chunk_text(extracted_text)
print(f"Created {len(chunks)} text chunks.")

# Step 4: Create embeddings and vector store (FAISS index)
def create_vector_store(chunks, model_name="all-MiniLM-L6-v2"):
    model = SentenceTransformer(model_name)
    embeddings = model.encode(chunks, convert_to_tensor=False)  # Create embeddings

    # Convert to numpy array (required by FAISS)
    embedding_array = np.array(embeddings).astype('float32')

    # Initialize FAISS index (using L2 distance)
    index = faiss.IndexFlatL2(embedding_array.shape[1])
    index.add(embedding_array)  # Add embeddings to FAISS index

    return index, embeddings, chunks

# Create the vector store
index, embeddings, chunks = create_vector_store(chunks)
print("Vector store created successfully.")

# Step 5: Query the vector store
def query_vector_store(index, query, chunks, model_name="all-MiniLM-L6-v2", k=3):
    model = SentenceTransformer(model_name)

    # Convert the query into an embedding
    query_embedding = model.encode([query], convert_to_tensor=False)

    # Perform a search in the FAISS index (retrieve top k results)
    distances, indices = index.search(np.array(query_embedding).astype('float32'), k)

    # Retrieve the most relevant chunks
    results = [chunks[i] for i in indices[0] if i < len(chunks)]
    return results

# Example query
query = "unemployment rate for bachelor's degree"
results = query_vector_store(index, query, chunks)
print("\nTop Matching Results:")
for idx, res in enumerate(results):
    print(f"{idx+1}. {res}")

```
